```python
import pandas as pd
df = pd.read_csv("meteo_api.csv")
df = df.loc[:, ~df.columns.str.contains("^Unnamed")]

# Fechas
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Solo Melbourne
df = df[df["Location"] == "Melbourne"]

# Desde 2008 hasta 2010 (finales)
mask = (df["Date"] >= "2008-01-01") & (df["Date"] <= "2010-12-31")
df = df[mask]

df.head()
```

```python
# Cogemos solo temperaturas máximas y minimas
df_small = df[[
    "Date",
    "Location",
    "temperature_2m_max",
    "temperature_2m_min"
]].sort_values("Date").reset_index(drop=True)

# Creamos el dataset nuevo
df_small.to_csv("melbourne2008-2010.csv", index=False)
df_small.head(), df_small.tail()
```
